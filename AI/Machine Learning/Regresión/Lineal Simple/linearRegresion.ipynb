# Importar Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

#-----------------IMPORT DATASET-----------------

dataset = pd.read_csv('/content/Notebooks/AI/Machine Learning/Regresión/Lineal Simple/Salary_Data.csv')
X = dataset.iloc[:,:-1].values 
Y = dataset.iloc[:,1].values
#--------------------ENTRENAMIENTO Y TEST----------------------

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 1/3, random_state = 0)
```

#Regresión Lineal
El modelo de regresión lineal simple, corresponde con la ecuación de la recta: 

$y = b_0 +b_1 \cdot x$

Donde: \\
$y$: variable dependiente \\
$b_0$: es la distancia el valor de la variable independiente cuando $x = 0$ \\
$b_1$: es la pendiente de la recta \\
$x$: variable independiente

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train, Y_train)
coefi = regression.fit(X_train, Y_train).coef_
print(np.size(coefi))
```

```python
#Predecir el conjunto de test
y_pred = regression.predict(X_test)

#Visualizar resultados de entrenamiento
plt.scatter(X_train,Y_train, color = "red")#graficar nube de puntos
    #recta de regresion lineal
plt.plot(X_train, regression.predict(X_train), color = "blue")
plt.title("Sueldo vs Años de Experiencia [conjunto de entrenamiento]")
plt.xlabel("Años de experiencia")
plt.ylabel("Sueldo (en $)")
plt.show()
```

```python
#Visualizar resultados de test
plt.scatter(X_test,Y_test, color = "red")#graficar nube de puntos
    #la recta de regresion lineal es la misma para cualquiera de los dos casos
plt.plot(X_train, regression.predict(X_train), color = "blue")
plt.title("Sueldo vs Años de Experiencia [conjunto de testing]")
plt.xlabel("Años de experiencia")
plt.ylabel("Sueldo (en $)")
plt.show()
```
