# Árboles de decisión para regresión

Classification And Regression Trees o CART por sus siglas en inglés. Es un método en el que se realizan divisiones por regiones, "hojas", tal que los datos se agrupen dentro de esta región.

Los árboles de decisión basan su proceso en el concepto de entropía, es decir, que los datos agrupados en un plano $X_1X_2$ representan una determinada característica común $y$. Considerando que los puntos dentro de dicho plano son una proyección de los puntos en un espacio 3-D, del sistema de coordenadas $X_1X_2Y$

Capture.PNG

El proceso consiste en agrupar los puntos por regiones, delimatadas por nodos hoja "Split", hasta que se obtenga la resolución deseada. Y procurando que se provea de suficiente información para que la classificación final tenga sentido.

Sin embargo, se procura evitar que los datos se dividan en "infinitas" regiones para prevenir problemas de overfitting. Por ejemplo, se puede especificar que cada hoja (región) tenga como mínumo un 5% del total de datos.

Para realizar las predicciones, se calcula la media de todos los puntos agrupados en cada región, y ese valor será el valor de predicción para dicha región (valores en verde de la gráfica). Por lo tanto, el valor de la predicción $y$ será el mismo para cualquier dato dentro de cada nodo hoja.

#Algoritmo

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 

#-----------------IMPORT DATASET-----------------

dataset = pd.read_csv('/content/Notebooks/AI/Machine Learning/Regresión/Árboles de desición/Position_Salaries.csv')
X = dataset.iloc[:,1:2].values 
y = dataset.iloc[:,2].values
#como el dataset es pequeño no tiene sentido separar en training y test
```

```python
#-------------AJUSTAR MODELO------------------
#*********obtener el modelo de regresión
from sklearn.tree import DecisionTreeRegressor
regression =  DecisionTreeRegressor(random_state = 0)# por defecto utiliza el error cuadrádico medio para definir los valores de corte
#https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeRegressor.html
regression.fit(X,y)
```

```python
#----------------PREDICCIÓN-------------------
regression.predict([[6.5]])
```

En la siguiente gráfica se puede apreciar como, para todos los valores dentro de los nodos hoja, el valor de predicción $y$ será el mismo.

```python
#plot
X_grid = np.arange(min(X),max(X),0.1)# para evitar que el gráfico salga a trozos
X_grid = X_grid.reshape(len(X_grid),1)
X_1 = X_grid #los X que se quieren predecir en el plot
plt.scatter(X, y, color = "red")
plt.plot(X_1, regression.predict(X_1), color = "blue")
plt.title("Modelo de Regresión Polinómico n4")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en $)")
plt.show()
```
